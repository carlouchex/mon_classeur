```python
phrase = 'Hello world'
majuscule = True
nouvelle_phrase = ''
for caractere in phrase:
    if majuscule:
        caractere = caractere.upper()
    else:
        caractere = caractere.lower()
    nouvelle_phrase += caractere
    majuscule = False

print(nouvelle_phrase)
```

```python
phrase = 'un dernier exercice et vous pourez vous reposer'
majuscule = True
nouvelle_phrase = ''
for caractere in phrase:
    if majuscule:
        caractere = caractere.upper()
        majuscule = False
    else:
        caractere = caractere.lower()
        majuscule = True
    nouvelle_phrase += caractere

print(nouvelle_phrase)
```

```python
nom = 'Melissa'
nouveau_nom = ''
for c in nom:
    if c == 's':
        c = '$'
    if c == 'e':
        c = '3'
    if c == 'a':
        c = '@'
    nouveau_nom += c

print(nouveau_nom)
```

```python
adn = 'AGTAGCGTC'
adn_inverse = ''
for caractere in adn:
    if caractere == 'A':
        caractere = 'T'
    else: 
        if caractere == 'T':
         caractere = 'A'
    if caractere == 'G':
        caractere = 'C'
    else:
        if caractere == 'C':
         caractere = 'G'
    adn_inverse += caractere

print(adn_inverse)
```

utilisation du mot elif= sinon si

```python
phrase = 'La touche e de mon clavier est cassée'
nouvelle_phrase = ''
for c in phrase:
    if c != 'e':
        nouvelle_phrase += c
print(nouvelle_phrase)
```

Le contraire de l'opérateur « est égal à » == est l'opérateur « est différent de » !=

```python
pourcentage = 73

if pourcentage < 40:
    note = 'F'
elif pourcentage < 60:
    note = 'C'
elif pourcentage < 80:
    note = 'B'
else:
    note = 'A'

print(note)
```

< (inférieur) et > (supérieur) sont deux autres opérateurs classiques. a < b signifie que « a est strictement inférieur à b
